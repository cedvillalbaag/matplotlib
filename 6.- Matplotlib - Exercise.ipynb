```python
# importar Librería 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set(style="darkgrid")
%matplotlib inline
```

```python
# Definir valores de X
x = np.linspace(-6,6,180)
```

```python
# Definir valores de Y
cos = np.cos(x)
sin = np.sin(x)
```

```python
# Graficar Subplot

# Definir Tamaño de Gráfico
plt.figure(figsize=(15,8))


#Definir Primer subplot
plt.subplot(221)
plt.plot(x, cos, color = 'red', linewidth = 2.6, linestyle= ':')
plt.title('cosx')
plt.grid(True)
plt.text(-5,-0.3, 'this will run across the subplot', rotation = 10, fontsize = 15)


#Definir Segundo Subplot
plt.subplot(222)
plt.plot(x, sin, color = 'blue', linewidth = 1.5, linestyle= '-.')
plt.title('sinx')
plt.grid(True)
plt.text(-5,-0.3, 'this will run across the subplot', rotation = 7, fontsize = 10)


#Tercer Subplot
plt.subplot(223)
plt.plot(x, -cos, color = 'green', linewidth = 1.5, linestyle= '-.')
plt.title('-cosx')
plt.grid(True)
plt.text(-5,-0.3, 'this will run across the subplot2', rotation = 7, fontsize = 10)

#Cuarto Subplot
plt.subplot(224)
plt.plot(x, -sin, color = 'orange', linewidth = 2, linestyle= '--')
plt.title('-sinx')
plt.grid(True)
plt.text(-5,-0.3, 'this will run across the subplot3', rotation = 7, fontsize = 10)
```
