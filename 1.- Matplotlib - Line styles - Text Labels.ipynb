```python
#Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
#from scipy.stats import spearmanr

#Dataset Seaborn
#sns.load_dataset("iris")
#sns.load_dataset("tips")
#sns.load_dataset("flights")

%matplotlib inline
```

```python
x = np.linspace(-6,6,200)
y = np.cos(x)
z = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,z)
```

```python
plt.plot(x,y)
plt.plot(x,z)
```

<h3> Line styles - Text Labels </h3>

```python
plt.plot(x,y, color = "green", linewidth=4, linestyle=":")
plt.plot(x,z)
```

```python
plt.plot(x,y, color = "orange", linewidth=4, linestyle="-.")
plt.plot(x,z)
```

```python
plt.plot(x,y, color = "orange", linewidth=4, linestyle="--")
plt.plot(x,z)
```

```python
plt.plot(x,y, color = "orange", linewidth=4, linestyle="-")
plt.plot(x,z)
```

```python
plt.plot(x,y, color = "orange", linewidth=4, linestyle="-")
plt.text(-4,1.25, "This text will be displayed across our plot", rotation = 0, fontsize= 16)
```

```python
plt.plot(x,y, color = "orange", linewidth=4, linestyle="-")
plt.text(-5,1.25, "This text will be displayed across our plot", rotation = 0, fontsize= 16)
```

```python
plt.subplot(211)
plt.plot(x,y, color = "orange", linewidth=4, linestyle="-")
plt.text(-5,1.25, "This text will be displayed across our plot", rotation = 0, fontsize= 10)

plt.subplot(212)
plt.plot(x, z, color = "red", linewidth=4, linestyle=":")
plt.text(-5,1.25, "This text will be displayed across our plot", rotation = 0, fontsize= 10)
```

```python
plt.subplot(121)
plt.plot(x,y, color = "orange", linewidth=4, linestyle="-")
plt.text(-5,1.25, "This text will be displayed across our plot", rotation = 0, fontsize= 10)

plt.subplot(122)
plt.plot(x, z, color = "red", linewidth=4, linestyle=":")
plt.text(-5,1.25, "This text will be displayed across our plot", rotation = 0, fontsize= 10)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(x,y, color = "orange", linewidth=4, linestyle="-")
plt.text(-5,1.25, "This text will be displayed across our plot", rotation = 0, fontsize= 10)

plt.subplot(122)
plt.plot(x, z, color = "red", linewidth=4, linestyle=":")
plt.text(-5,1.25, "This text will be displayed across our plot", rotation = 0, fontsize= 10)
```
