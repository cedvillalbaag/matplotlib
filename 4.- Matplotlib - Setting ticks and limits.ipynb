```python
#Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set(style="darkgrid")
#from scipy.stats import spearmanr

#Dataset Seaborn
#sns.load_dataset("iris")
#sns.load_dataset("tips")
#sns.load_dataset("flights")

%matplotlib inline
```

```python
x = np.linspace(-6,6,225)
y = np.cos(x)
z = np.sin(x)
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y, color = "orange", linewidth=4, linestyle="--")
plt.plot(x,z)

plt.xlabel("Eje x", fontsize=20, color = "blue")
plt.ylabel("Eje y", fontsize=20, color= "red")
plt.title("Ejemplo de Gráfica")

#escala de eje x
plt.xticks(np.linspace(-6,6,12))
#escala de eje y
plt.yticks(np.linspace(-1,1,6))

#limite de eje x
plt.xlim(-9,9)
#limite de eje y
plt.ylim(-2,2)
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y, color = "orange", linewidth=4, linestyle="--")
plt.plot(x,z)

plt.xlabel("Eje x", fontsize=20, color = "blue")
plt.ylabel("Eje y", fontsize=20, color= "red")
plt.title("Ejemplo de Gráfica")

#escala de eje x
plt.xticks(np.linspace(-6,6,12))
#escala de eje y
plt.yticks(np.linspace(-1,1,6))

#limite de eje x
plt.xlim(x.min()-2, x.max()+2)
#limite de eje y
plt.ylim(y.min()-2, y.max()+2)
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y, color = "orange", linewidth=4, linestyle="--")
plt.plot(x,z)

plt.xlabel("Eje x", fontsize=20, color = "blue")
plt.ylabel("Eje y", fontsize=20, color= "red")
plt.title("Ejemplo de Gráfica")
```
