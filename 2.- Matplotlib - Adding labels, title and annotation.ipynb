```python
#Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set(style="darkgrid")
#from scipy.stats import spearmanr

#Dataset Seaborn
#sns.load_dataset("iris")
#sns.load_dataset("tips")
#sns.load_dataset("flights")

%matplotlib inline
```

```python
x = np.linspace(-10,10,150)
y = np.cos(x)
z = np.sin(x)
```

```python
plt.plot(x,y, color = "orange", linewidth=4, linestyle="--")
plt.plot(x,z)
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Ejemplo de Gráfica")
```

```python
plt.plot(x,y, color = "orange", linewidth=4, linestyle="--")
plt.plot(x,z)
plt.xlabel("Eje x", fontsize=20, color = "blue")
plt.ylabel("Eje y", fontsize=20, color= "red")
plt.title("Ejemplo de Gráfica")
```

```python
plt.plot(x,y, color = "orange", linewidth=4, linestyle="--")
plt.plot(x,z)
plt.xlabel("Eje x", fontsize=20, color = "blue")
plt.ylabel("Eje y", fontsize=20, color= "red")
plt.title("Ejemplo de Gráfica")

plt.annotate("este es el valor máximo", xy=(0,1), xytext=(10,1.2), arrowprops=dict(facecolor="green", shrink=0.001))
```
