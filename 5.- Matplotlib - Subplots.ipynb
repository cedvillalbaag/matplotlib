### Subplot
Ejemplo: subplot(211) ó Subplot(2,1,1)
 
Lista de Elementos:
- 2: Numero de filas
-  1: Numero de columnas
-  1: Numero de figura 


```python
x = np.linspace(-10,10,150)
y = np.cos(x)
z = np.sin(x)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(211)
plt.plot(x,y, color = "orange", linewidth=4, linestyle="--")


plt.subplot(212)
plt.plot(x,z)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(121) #Una fila , Dos columna, Primera figura
plt.plot(x,y, color = "orange", linewidth=4, linestyle="--")


plt.subplot(122)
plt.plot(x,z) #Una fila, dos columnas, segunda figura
```

```python

plt.axes([-0.4,1,1,1]) #left,bottom,width,height
plt.plot(x,y, color = "orange", linewidth=4, linestyle="--")


plt.axes([1,1,1,1])
plt.plot(x,z)
```

```python
plt.axes([-0.4,1,1,1]) #left,bottom,width,height
plt.plot(x,y, color = "orange", linewidth=4, linestyle="--")


plt.axes([1,1,1,0.5])#left,bottom,width,height
plt.plot(x,z)
```
